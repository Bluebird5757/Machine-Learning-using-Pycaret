```python
setup(data = myDataSet, target = 'medv',
      feature_selection = True,
      feature_selection_method = 'univariate',
      n_features_to_select = 0.2,
      verbose = False)

cm = compare_models()
```

```python
setup(data = myDataSet, target = 'medv',
      remove_outliers = True,
      outliers_method = "iforest",
      outliers_threshold = 0.05,
      verbose = False)

cm = compare_models()
```
